1. Find prime factors
Ek number ke saare prime factors nikalne ke liye:

```python
def find_prime_factors(n):
    factors = []
    # 2 se divide karte hain jab tak ho sake
    while n % 2 == 0:
        factors.append(2)
        n //= 2
    # Odd numbers check karte hain
    for i in range(3, int(n**0.5) + 1, 2):
        while n % i == 0:
            factors.append(i)
            n //= i
    if n > 2:
        factors.append(n)
    return factors

find_prime_factors(25)
```

2. Identify a palindrome
String ulti aur seedhi dono taraf se same hai ya nahi check karne ke liye:

```python
def is_palindrome(s):
    # Spaces hatakar lowercase mein convert karte hain
    clean_s = "".join(s.split()).lower()
    return clean_s == clean_s[::-1]

(is_palindrome("51 21215 "))  
```

3. Sort a string

```python
def sort_string(s):
    return "".join(sorted(s))

print(sort_string("jdkjuejkbdhlfuwhe"))  # Output: hhnopty
```

4. Find all list items
Kisi list mein kisi specific item ke saare index positions dhoondhne ke liye:

```python
def find_all_items(lst, item):
    return [index for index, value in enumerate(lst) if value == item]

fruits = ["apple", "banana", "apple", "cherry", "apple"]
type('print(find_all_items(fruits, "kela"))')
```

7. Schedule a function
Kisi function ko kuch seconds ke baad run karne ke liye sched module ka use:

```python
import sched; import time

scheduler = sched.scheduler(time.time, time.sleep)

def print_message(name):
    print(f"Hello {name}, function chal gaya!")

# 3 seconds ke baad run hoga (Priority 1)
scheduler.enter(3, 1, print_message, argument=("Amit",))
print("Scheduler shuru ho raha hai...")
scheduler.run()
```

9. Simulate dice
Ludo ka paasa (Dice) roll karne ke liye

```python
import random

def roll_dice(num_dice=1):
    return [random.randint(1, 6) for _ in range(num_dice)]

print(roll_dice(2))  # Output: Jaise [4, 6]
```

10. Count unique words
Kisi paragraph ya string mein unique words aur unka count nikalne ke liye:

```python
def count_unique_words(text):
    words = text.lower().split()
    # Punctuation hatane ke liye basic safai
    clean_words = [w.strip(".,!?;:") for w in words]
    
    word_counts = {}
    for word in clean_words:
        if word:
            word_counts[word] = word_counts.get(word, 0) + 1
    return word_counts

text = "Apple orange apple banana orange apple"
print(count_unique_words(text)) # Output: {'apple': 3, 'orange': 2, 'banana': 1}
```

11. Generate a password
Ek strong random password generate karne ke liye:

```python
import string
import random

def generate_password(length=12):
    characters = string.ascii_letters + string.digits + string.punctuation
    password = "".join(random.choice(characters) for _ in range(length))
    return password

print(generate_password(16))  # Output: Jaise 'kP9@mQ2!xZ9#rT4$'
```

12. Merge CSV files
Multiple CSV files ka data ek hi file mein merge karne ke liye:

```python
import pandas as pd
import glob

def merge_csv_files(output_filename="merged.csv"):
    # Saari CSV files ki list banate hain
    csv_files = glob.glob("*.csv") 
    
    # Agar files hain toh merge karenge
    if csv_files:
        combined_df = pd.concat([pd.read_csv(f) for f in csv_files], ignore_index=True)
        combined_df.to_csv(output_filename, index=False)
        print(f"{len(csv_files)} files merge ho gayi hain!")
    else:
        print("Koi CSV file nahi mili.")
```

13. Solve a sudoku
Sudoku board ko solve karne ke liye Backtracking Algorithm:

```python
def is_valid(board, row, col, num):
    # Row check
    for x in range(9):
        if board[row][x] == num:
            return False
    # Column check
    for x in range(9):
        if board[x][col] == num:
            return False
    # 3x3 Box check
    start_row, start_col = row - row % 3, col - col % 3
    for i in range(3):
        for j in range(3):
            if board[i + start_row][j + start_col] == num:
                return False
    return True

def solve_sudoku(board):
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:  # Khali jagah
                for num in range(1, 10):
                    if is_valid(board, i, j, num):
                        board[i][j] = num
                        if solve_sudoku(board):
                            return True
                        board[i][j] = 0  # Backtrack
                return False
    return True
```
